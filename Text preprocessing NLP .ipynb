# Text preprocessing

# Read the data

```python
# Read the data. Here it is already in .csv format.
import pandas as pd
data = pd.read_csv('SPAM text message 20170820 - Data.csv')
data.head()
```

```python
pd.set_option("display.max_colwidth",100)
data = pd.read_csv('SPAM text message 20170820 - Data.csv')
data.head()
```

# Remove puncuations

```python
import string as st
def remove_punct(text):
    return ("".join([c for c in text if c not in st.punctuation]))
```

```python
data['removed_punc'] = data['Message'].apply(lambda x: remove_punct(x))
data.head()
```

# Tokenization

```python
import re

def tokenize(text):
    text = re.split('\W+' ,text)
    return [x.lower() for x in text]
```

```python
data['tokens'] = data['removed_punc'].apply(lambda msg : tokenize(msg))
data.head()
```

# Remove stop words

```python
import nltk
stopwords=nltk.corpus.stopwords.words('english')
stopwords[0:10]
```

```python
import nltk

def remove_stopwords(text):
    return [word for word in text if word not in nltk.corpus.stopwords.words('english')]
```

```python
data['stop_words'] = data['tokens'].apply(lambda x : remove_stopwords(x))
data.head()
```

# Stemming

```python
from nltk.stem import PorterStemmer
def stemming(text):
    ps = PorterStemmer()
    return [ps.stem(word) for word in text]
```

```python
data['stem_words'] = data['stop_words'].apply(lambda wrd: stemming(wrd))
data.head()
```

# Lemmatization

```python
from nltk import WordNetLemmatizer
# Apply lemmatization on tokens
def lemmatize(text):
    word_net = WordNetLemmatizer()
    return [word_net.lemmatize(word) for word in text]
```

```python
data['lemma_words'] = data['stop_words'].apply(lambda x : lemmatize(x))
data.head()
```

# Return sentence

```python
# Create sentences to get clean text as input for vectors

def return_sentences(tokens):
    return " ".join([word for word in tokens])
```

```python
data['clean_text'] = data['lemma_words'].apply(lambda x : return_sentences(x))
data.head()
```
